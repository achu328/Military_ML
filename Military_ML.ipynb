```python
import zipfile

file_path = "/content/drive/MyDrive/Military.zip"
Output_folder= "/content/drive/MyDrive/Corrected/yolov5/data_yolo/data"

# Extract zip
with zipfile.ZipFile(file_path, 'r') as zip_ref:
    zip_ref.extractall(Output_folder)

print("ZIP file extracted successfully.")
```

```python
# Create a Parameter for Yolov5 and Yaml:
para={

        "train":str(Path("/content/drive/MyDrive/Corrected/yolov5/data_yolo/data/military_object_dataset/train/images")),
        "val":str(Path("/content/drive/MyDrive/Corrected/yolov5/data_yolo/data/military_object_dataset/val/images")),
        "test":str(Path("/content/drive/MyDrive/Corrected/yolov5/data_yolo/data/military_object_dataset/test/images")),
        "nc": 12,
        "names": ['Marksman', 'Weapon', 'Tank', 'War Truck', 'Jeep', 'Ground','Soldier','Helicopter', 'Missile', 'Radar', 'War Plane', 'Warship']

}

with open("/content/drive/MyDrive/Corrected/yolov5/data_yolo/train_data.yaml", "w") as train_config:
    yaml.dump(para, train_config)

print("train_data.yaml created.")
```
